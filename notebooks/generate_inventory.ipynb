```python
import pandas as pd
import random
```

```python
# 1. 기초 데이터 설정 (아모레퍼시픽 브랜드 및 제품 예시)
brands = ["설화수", "아이오페", "라네즈", "마몽드", "에스트라", "한율"]
product_types = ["에센스", "크림", "앰플", "토너", "선크림", "마스크팩"]
stock_statuses = ["재고 충분", "재고 적정", "재고 부족"]
skin_types = ["건성", "지성", "복합성", "민감성"]

# 상세 제품 정보 데이터베이스 (예시)
raw_data = [
    ["윤조에센스", "설화수", "에센스", "자음단", "피부 자생력 강화 및 윤기 케어", "모든 피부", "탄력 및 윤기"],
    ["슈퍼바이탈 크림", "아이오페", "크림", "슈퍼플라보노이드", "항노화 및 고영양 공급", "건성", "안티에이징"],
    ["워터뱅크 블루 히알루로닉", "라네즈", "크림", "블루 히알루론산", "초저분자 수분 공급", "복합성", "깊은 보습"],
    ["바쿠치올 레티놀 크림", "마몽드", "크림", "바쿠치올", "저자극 탄력 케어", "민감성", "주름 개선"],
    ["아토바리어 365 크림", "에스트라", "크림", "더마온(세라마이드)", "피부 장벽 강화", "민감성", "장벽 보호"],
    ["어린쑥 진정수", "한율", "토너", "어린쑥 추출물", "피부 진정 및 수분 공급", "모든 피부", "진정 케어"],
    ["레티놀 엑스퍼트", "아이오페", "에센스", "순수 레티놀", "집중 주름 개선", "지성/복합성", "모공 및 주름"],
    ["자음생 크림", "설화수", "크림", "인삼 결정체", "피부 자생 및 활력", "건성/노화", "영양 탄력"],
    ["워터 슬리핑 마스크", "라네즈", "마스크팩", "슬립톡스", "밤사이 집중 수분 충전", "모든 피부", "수분 광채"],
    ["에이시카 스트레스 릴리프", "에스트라", "앰플", "CICA분석물", "급속 진정", "민감성", "트러블 진정"]
]

# 2. 데이터프레임 생성
inventory_list = []

# 데이터 양을 늘리기 위해 반복 및 랜덤 재고 할당
for i in range(50):  # 50개의 재고 행 생성
    base = random.choice(raw_data)
    inventory_list.append({
        "제품명": base[0],
        "브랜드": base[1],
        "제품유형": base[2],
        "핵심성분": base[3],
        "성분특징": base[4],
        "재고상태": random.choice(stock_statuses), # 실시간성을 위해 랜덤 부여
        "추천피부": base[5],
        "타입효과": base[6]
    })

df_inventory = pd.DataFrame(inventory_list)

# 3. CSV 파일로 저장
file_name = "amore_inventory.csv"
df_inventory.to_csv(file_name, index=False, encoding='utf-8-sig')

print(f"'{file_name}' 파일이 성공적으로 생성되었습니다.")
print(df_inventory.head())
```
